```python
import torch
import numpy as np
# tensor  
x = torch.rand(4,3)
print(x)
y =torch.randn(4,3)
print(y)
```

```python
#  
a = torch.zeros((4,4),dtype=torch.long)
print(a)
# 
b = torch.ones(4,4)
print(b)
c = torch.tensor(np.ones((2,3),dtype='int32'))
print(c)
```

Tensor


```python
# tensor 
# 
print(a+b)
# add_ = replace in 
y = a.add_(3)
print(y)
```

```python
# torch.view()torch.reshape()
x = torch.randn(4,3)
y = x.view(12)
z = x.view(-1,6)
print(x.size(),y.size(),z.size())
print(x)
print(z)
# x tensor size 
#view()
print(x)
```

```python
#
#torch.reshape() 
#
a = torch.randn(4,4)
b = a.reshape(2,8)
print(a)
print(b)
```

```python
#
# Tensor (broadcasting) Tensor 
x = torch.arange(1,4).view(1,3)
print(x)
y = torch.arange(1,5).view(4,1)
print(y)
print(x+y)
```
